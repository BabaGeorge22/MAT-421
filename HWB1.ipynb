Section 17.1 – Interpolation Problem Statement

We have a set of points
(xi, yi). We want a function that goes through every point exactly. This is called interpolation. We then use this function to estimate new
y-values for new
x-values.

```python
import matplotlib.pyplot as plt

# Data
x_data = [0, 1, 2, 3]
y_data = [2, 3, 2.5, 4]

# Plot the data
plt.figure(figsize=(6,4))
plt.plot(x_data, y_data, 'o')
plt.title("Interpolation Concept")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Import matplotlib.pyplot to make a plot.

Set up sample data (x_data, y_data) and plot it.

The plot shows how points might look before we do any interpolation.

Section 17.2 – Linear Interpolation

Linear interpolation connects each pair of data points with a straight line.

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

# Data for linear interpolation
x_vals = [2, 4, 6]
y_vals = [1, 6, 4]

# Linear interpolation function
lin_f = interp1d(x_vals, y_vals, kind='linear')

# Interpolate at a new x
x_new = 3
y_new = lin_f(x_new)
print("Linear interpolation at x=3 ->", y_new)

# Plot
plt.figure(figsize=(6,4))
plt.plot(x_vals, y_vals, '-ob')
plt.plot(x_new, y_new, 'ro')
plt.title("Linear Interpolation at x = 3")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Built a linear function with interp1d and tested it at x=3.

Plotted everything with a blue line and a red square for the interpolated point.

Section 17.3 – Cubic Spline Interpolation

A cubic spline uses smooth, cubic curves for each pair of points, making the interpolation smooth at all points.

```python
from scipy.interpolate import CubicSpline

# Data points for cubic spline
x_data = [0, 2, 4]
y_data = [1, 3, -1]

# The cubic spline
c_spline = CubicSpline(x_data, y_data, bc_type='natural')

# Points to see the smooth curve
x_fine = np.linspace(0, 4, 100)
y_fine = c_spline(x_fine)

# Interpolate at x=1.0
x_star = 1.0
y_star = c_spline(x_star)
print("Cubic spline at x=1.0 ->", y_star)

# Plot
plt.figure(figsize=(6,4))
plt.plot(x_data, y_data, 'ro', label='Data Points')
plt.plot(x_fine, y_fine, 'b', label='Cubic Spline')
plt.plot(x_star, y_star, 'go', label='Interpolated')
plt.title("Cubic Spline Interpolation")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Used x=[0,2,4], y=[1,3,-1] for cubic splines.

Created a smooth function that fits these points.

I picked x=1 to see the interpolated y-value.

Plotted the original points and the smooth spline.
